```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from collections import Counter
```

## Load Data

```python
DATA_FILE = '../data/avspeech_test_langs.csv'
```

```python
data = pd.read_csv(DATA_FILE)
data
```

## Plotting Helpers

```python
def plot_pie(data_list, title=None):
    counts = Counter(data_list)
    labels, counts = zip(*counts.items())
    plt.pie(counts, labels=labels)
    plt.title(title)
```

## Data Analysis

```python
plot_pie(data['Language'], title="Language Distribution")
```
